```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('iris.csv').iloc[:,1:]
df
```

```python
X = df.drop(columns= ['Species'])
X
```

```python
y = df.Species.str.strip().eq('Iris-setosa').astype(int)
y
```

```python
y.unique()
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    X,y,test_size= 0.2,stratify = y, random_state =42
)
```

```python
y_train.value_counts(normalize = True)
```

```python
y_test.value_counts(normalize = True)
```

```python
model = LogisticRegression().fit(x_train,y_train)
```

```python
y_test_pred = model.predict(x_test)
y_train_pred = model.predict(x_train)
```

```python
y_train_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(f"test_accuracy: {accuracy_score(y_test,y_test_pred)}")
print(f"train_accuracy: {accuracy_score(y_train,y_train_pred)}")

print(f"test_classification_report: {classification_report(y_test,y_test_pred)}")
print(f"train_classification_report: {classification_report(y_train,y_train_pred)}")
```
